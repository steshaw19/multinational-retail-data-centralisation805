```python
import json
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def download_s3_data(s3_url):
    response = requests.get(s3_url)
    date_data = json.loads(response.text)
    return date_data

date_time_data = download_s3_data('https://data-handling-public.s3.eu-west-1.amazonaws.com/date_details.json')
```

```python
date_time_df= pd.DataFrame(date_time_data)
```
